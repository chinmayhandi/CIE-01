```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('/content/drive/MyDrive/AIML DATASETS /mtcars.csv')
df.head()
```

```python
auto=(df['am'] == 0).sum()
manual=(df['am'] == 1).sum()
plt.bar(['Automatic','Manual'],[auto,manual], color=['Red','blue'])
plt.title('Transmission')
plt.xlabel('types of Transmission')
plt.ylabel('Number of cars')
plt.show()
```

```python
plt.scatter(df['wt'], df['mpg'], color='green')
plt.title('Weight vs MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```

```python
auto=(df['am'] == 0).sum()
manual=(df['am'] == 1).sum()
plt.bar(['Automatic','Manual'],[auto,manual], color=['yellow','blue'])
plt.title('Transmission')
plt.xlabel('types of Transmission')
plt.ylabel('Number of cars')
plt.show()
```

```python
plt.boxplot(df['mpg'])
plt.title('MPG Boxplot')
plt.ylabel('Mpg')
plt.show()
```
