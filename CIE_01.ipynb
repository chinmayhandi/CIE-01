```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('/content/mtcars.csv')
df
```

```python
plt.hist(df['mpg'], bins=8 , color='red', edgecolor='black')
plt.title("MPG cars ")
plt.xlabel("mpg")
plt.ylabel("Count")
plt.show()
```

```python

plt.scatter(df['wt'], df['mpg'], color='green')
plt.title("Car weigth and mpg")
plt.xlabel("wt")
plt.ylabel("mpg")
plt.show()
```

```python
auto=(df['am']==0).sum()
manual=(df['am']==1).sum()
plt.bar(['automatic', 'manual' ], [auto, manual], color=['orange', 'blue'])
plt.title("type of cars ")
plt.xlabel("Types of cars ")
plt.ylabel("Count")
plt.show()
```

```python
plt.boxplot(df['mpg'])
plt.title("MPG cars ")
plt.xlabel("mpg")
plt.show()
```
